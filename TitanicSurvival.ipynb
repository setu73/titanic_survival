```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
titanic=sns.load_dataset('titanic')
titanic
```

```python
titanic.drop(titanic.iloc[:, 8:14], inplace=True, axis=1)
titanic
```

```python
onehot_titanic = pd.get_dummies(titanic, columns = ['pclass', 'sex', 'embarked'])
print(onehot_titanic)
```

```python
onehot_titanic['age'] = onehot_titanic['age'].replace(np.nan, onehot_titanic['age'].mean())
print(onehot_titanic)
```

```python
onehot_titanic = onehot_titanic[['age','sibsp','parch','fare','pclass_1','pclass_2','pclass_3', 'sex_female','sex_male','embarked_C','embarked_Q','embarked_S','survived']]
```

```python
print(onehot_titanic)
```

```python
X=onehot_titanic.iloc[:,0:12]
column_data_types = X.dtypes

print("Data types of all columns:")
print(column_data_types)
X
```

```python
Y=onehot_titanic.iloc[:,-1]
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.125, random_state=42)
```

```python
type(X_train)
```

```python
from keras import Model, Input
from keras.layers import Dense

inp = Input(shape=(12,))
l1 = Dense(50, activation='sigmoid')(inp)
out = Dense(1, activation='sigmoid')(l1)
model = Model(inputs=inp, outputs=out)
model.summary()
```

```python
model.compile(optimizer='adam', loss='mse')
model.fit(x=X_train, y=y_train, epochs=200, shuffle=True)
```

```python
from sklearn.metrics import accuracy_score

y_test=np.array(y_test)
X_test=np.array(X_test)

Y_pred1=model.predict(X_test).round()


score = accuracy_score(y_test, Y_pred1, normalize=True)
print("Accuracy of test data: ", score)
```

```python
X_train = np.array(X_train)
y_train = np.array(y_train)

Y_pred2=model.predict(X_train).round()

score = accuracy_score(y_train, Y_pred2, normalize=True)
print("Accuracy of training data: ", score)
```

```python
from keras.models import Model
from keras.layers import Input, Dense
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score

def create_model():
    inp = Input(shape=(12,))
    l1 = Dense(8, activation='sigmoid')(inp)
    out = Dense(1, activation='sigmoid')(l1)
    model = Model(inputs=inp, outputs=out)
    model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
    return model

model = KerasClassifier(model=create_model, epochs=10, batch_size=32, verbose=0)
scores = cross_val_score(model, X, Y, cv=5, scoring='accuracy')
print("Cross-validation scores:", scores)
print("Mean score:", scores.mean())
print("Standard deviation:", scores.std())
```

The standard deviation is not too large, which suggests that the model is not overfitting.
